```python
vaccination = df[df.State!='India']
vaccination
```

```python
First_Dose = vaccination.groupby('State')['First Dose Administered'].sum().to_frame('Total First Dose Administered')
First_Dose
```

```python
Second_Dose = vaccination.groupby('State')['Second Dose Administered'].sum().to_frame('Total Second Dose Administered')
Second_Dose
```
